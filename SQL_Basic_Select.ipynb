<p>The following practice problems were completed on <a href="www.hackerrank.com">HackerRank.com</a>.</p>
<p>Generally speaking, the problems ascend from easier to more difficult.</p>
<p>Relational Database Management System: <b>MySQL</b></p>

## Problem Set Table #1

The CITY table is described as follows:
<img src='images/1449729804-f21d187d0f-CITY.jpg'>

### Problems
1. Select All: Query all columns (attributes) for every row in the CITY table. (Difficulty: <span style='color:green'>Easy</span>)
2. Select By ID: Query all columns for a city in CITY with the ID 1661. (Difficulty: <span style='color:green'>Easy</span>)
3. Japanese Cities' Attributes: Query all attributes of every Japanese city in the CITY table. The COUNTRYCODE for Japan is JPN. (Difficulty: <span style='color:green'>Easy</span>)
4. Japanese Cities' Names: Query the names of all the Japanese cities in the CITY table. The COUNTRYCODE for Japan is JPN. (Difficulty: <span style='color:green'>Easy</span>)

#### Query #1
Select All: Query all columns (attributes) for every row in the CITY table. (Difficulty: <span style='color:green'>Easy</span>)

SELECT *

#### Query #2
Select By ID: Query all columns for a city in CITY with the ID 1661. (Difficulty: <span style='color:green'>Easy</span>)
##### Query
SELECT *  
FROM CITY  
WHERE ID = 1661;
##### Output
1661 Sayama JPN Saitama 162472 

#### Query #3
Japanese Cities' Attributes: Query all attributes of every Japanese city in the CITY table. The COUNTRYCODE for Japan is JPN. (Difficulty: <span style='color:green'>Easy</span>)
##### Query
SELECT *  
FROM CITY  
WHERE COUNTRYCODE = 'JPN';
##### Output

Japanese Cities' Names: Query the names of all the Japanese cities in the CITY table. The COUNTRYCODE for Japan is JPN. (Difficulty: <span style='color:green'>Easy</span>)

<img src='images/1449345840-5f0a551030-Station.jpg'>

### Problems
1. Weather Observation Station 1: Query a list of CITY and STATE from the STATION table. (Difficulty: <span style='color:green'>Easy</span>)
2. Weather Observation Station 3: Query a list of CITY names from STATION with even ID numbers only. You may print the results in any order, but must exclude duplicates from your answer. (Difficulty: <span style='color:green'>Easy</span>)
3. Weather Observation Station 4: Let N be the number of CITY entries in STATION, and let N' be the number of distinct CITY names in STATION; query the value of N - N' from STATION. 4. Weather Observation Station 5: Query the two cities in STATION with the shortest and longest CITY names, as well as their respective lengths (i.e.: number of characters in the name). 6. Weather Observation Station 7: Query the list of CITY names ending with vowels (a, e, i, o, u) from STATION. Your result cannot contain duplicates. (Difficulty: <span style='color:green'>Easy</span>)
7. Weather Observation Station 8: Query the list of CITY names from STATION which have vowels (i.e., a, e, i, o, and u) as both their first and last characters. Your result cannot contain duplicates. (Difficulty: <span style='color:green'>Easy</span>)
8. Weather Observation Station 9: Query the list of CITY names from STATION that do not start with vowels. Your result cannot contain duplicates. (Difficulty: <span style='color:green'>Easy</span>)
9. Weather Observation Station 10: Query the list of CITY names from STATION that do not end with vowels. Your result cannot contain duplicates. (Difficulty: <span style='color:green'>Easy</span>)
10. Weather Observation Station 11: Query the list of CITY names from STATION that either do not start with vowels or do not end with vowels. Your result cannot contain duplicates. (Difficulty: <span style='color:green'>Easy</span>)
11. Weather Observation Station 12: Query the list of CITY names from STATION that do not start with vowels and do not end with vowels. Your result cannot contain duplicates. (Difficulty: <span style='color:green'>Easy</span>)
